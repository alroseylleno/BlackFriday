#### Male & Female counts based on Age 

```python
data_gender=data.loc[data['Gender'].isin(['F','M'])]
data_gender_plot1=data[data['Gender']=="F"].Age.value_counts()
data_gender_plot2=data[data['Gender']=="M"].Age.value_counts()
#print('Female:',data_gender_plot1,'Male:',data_gender_plot2)
#data_gender_plot1.plot(kind="barh",figsize=(10,2),title='Female')
#data_gender_plot2.plot(kind="barh",figsize=(10,2),title='Male')
plt.plot(data_gender_plot1) 
plt.plot(data_gender_plot2)
plt.show()
```

```python
data_city_cat=data.loc[(data['City_Category'].isin(['A','B','C']))&(data['Gender'])]
data_city_cat
```
